#알라딘 API 사용해 제목, isbn 데이터 저장
베스트셀러의 isbn데이터 : query url을 베스트셀러로 100개씩 최대 5000번 요청 가능 : 500000 권

```python
import requests
import json

TTBKey = "ttbleejunho09091938001"
book_lists = []

for i in range(0,500):
  url = f"http://www.aladin.co.kr/ttb/api/ItemList.aspx?ttbkey={TTBKey}&QueryType=Bestseller&MaxResults=100&start={i}&SearchTarget=Book&output=JS&Version=20131101"
  res = requests.get(url)
  res.raise_for_status()

  for item in res.json()['item']:
    book_lists.append(item['isbn13'])
```

```python
print(len(book_lists))
```

```python
import csv

#csv로 저장하기
with open('isbn13_list.csv','w',newline = '', encoding='utf-8') as csvfile:
  writer = csv.writer(csvfile)

  for isbn in book_lists:
    writer.writerow([isbn])

print('저장왈료')
```

#웹 크롤링으로 상품 제목, 분야, 설명, 리뷰 등 가져오기

```python
from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
#csv 데이터 불러오기
isbn13_list_raw = []

with open('isbn13_list.csv','r',encoding='utf-8') as csvfile:
  reader = csv.reader(csvfile)
  for row in reader:
    isbn13_list_raw.append(row)
```

```python
isbn13 = isbn13_list_raw[100][0]

url =  f'https://www.aladin.co.kr/shop/wproduct.aspx?ISBN={isbn13}'
print(url)

res = requests.get(url)
res.raise_for_status()
soup = BeautifulSoup(res.text,'lxml')
```

```python
#제목
title = soup.find("div", attrs={"class":"title", "class":"layer_tlist"}).ul.li.a.get_text()
#부가설명
sub_title = soup.find("div", attrs={"class":"sub_title", "class":"layer_tlist"}).ul.li.get_text()

print(title, sub_title)
```
