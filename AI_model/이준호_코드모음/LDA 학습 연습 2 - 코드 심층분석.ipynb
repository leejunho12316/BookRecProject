```python
from sklearn.datasets import fetch_20newsgroups
import pandas as pd
```

```python
dataset = fetch_20newsgroups(remove = ('headers','footers','quotes'))
docs = dataset.data
targets = dataset.target
all_targets = dataset.target_names
```

```python
df = pd.DataFrame({'document':docs})
df['clean_str'] = df['document'].str.replace("[^a-zA-Z]", " ", regex=True)
df['only_words'] = df['clean_str'].apply(lambda x : ' '.join(w for w in x.split() if len(w) > 3))
df['clean'] = df['only_words'].apply(lambda x : ''.join(x.lower()))
```

```python
df
```

```python
import nltk
nltk.download('stopwords')
```

```python
stop_words = nltk.corpus.stopwords.words('english')
print(stop_words)
```

```python
tokenized_doc = df['clean'].apply(lambda x : [word for word in x.split() if word not in stop_words] )
tokenized_doc
```

```python
from gensim import corpora
dictionary = corpora.Dictionary(tokenized_doc)
corpus = [dictionary.doc2bow(text) for text in tokenized_doc]
```

```python
import gensim

ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics = 20, id2word = dictionary, passes=15)
```
